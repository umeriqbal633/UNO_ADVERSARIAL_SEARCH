## 2. Minimax vs. Expectimax - Strategy Employment

### **Minimax (Defensive)**
Minimax assumes the worst-case scenario. It treats the opponents' actions as optimal attempts to actively minimize the AI's score. Because 3-player UNO is a multi-agent environment where enemies are not perfectly aligned against the AI, assuming they play perfectly cooperatively against Player 1 is a safe, pessimistic, and robust approach.

### **Expectimax (Offensive)**
Expectimax treats "Drawing" from the deck as a **Chance Node** rather than a strict opponent move. It takes into account the underlying probability distribution of cards remaining in the deck (colors and values) to calculate an expected value for the `Draw` action.
- It treats opponent turns as *stochastic nodes* (calculating the average expected score across all valid moves an opponent could make) rather than assuming opponents actively sabotage Player 2. This lack of pessimism allows the agent to make riskier, more aggressive plays.

## 3. Algorithm Performance Comparison

### **Which algorithm performed best?**
Over repeated simulated runs, **Expectimax (Offensive)** generally demonstrates superior win rates and faster card clearing behaviors compared to standard Minimax.

### **Reasoning behind superior performance**
1. **Nature of the Game**: UNO is largely a stochastic game based closely on chance (hidden hands and masked deck pulls) rather than a deterministic game of perfect information like Chess. 
2. **Over-pessimism in Minimax**: Minimax assumes that opponents will universally make the absolute worst-case move for the AI. In a 3-player UNO game, this is rarely true—Player 2 and Player 3 are often busy attacking each other, not solely aiming to destroy Player 1. Minimax ends up playing too safely.
3. **Predictive Deck State in Expectimax**: By treating the draw system as a probability-based Chance node, Expectimax can deduce that if 75% of the remaining deck is 'Red', it might aggressively play a card to switch the color to 'Red', knowing its expected future outcomes are superior. It maximizes raw statistical advantage, out-scaling Minimax's defensive passivity.
